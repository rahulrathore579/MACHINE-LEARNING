```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split          # for selecting and training
from sklearn.linear_model import LinearRegression             #
from sklearn.metrics import mean_squared_error, r2_score      # fro evaluation we use sklearn.metrics

data=pd.read_csv('/content/aircraft_fuel_consumption.csv')
data
```

```python
x1=data['Flight_Distance']
x2=data['Number_of_Passengers']
x3=data['Flight_Duration']
x4=data['Aircraft_Type']    # it is categorical data and some model is not accept in modeling os we convert it into numerical data by encoding
from sklearn.preprocessing import LabelEncoder


'''tupe of encoding
label encoding
one hot encoding
binary encoding
frequency encoding
ordinal encoding'''

#x=data[[x1,x2,x3,x4]]      x variable m multipule variable store nii hota

X=data[['Flight_Distance','Number_of_Passengers','Flight_Duration','Aircraft_Type']]
Y=data['Fuel_Consumption']
X=pd.get_dummies(X,columns=['Aircraft_Type'])
X.columns
X
```

```python
# split the data into traing and testing
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# initiallization and train model
model=LinearRegression()
model.fit(X_train,Y_train)

#model prediction
y_pred = model.predict(X_test)
print(y_pred)
#print(y_pred,X_test)
```

```python
# claculate mse
mse= mean_squared_error(Y_test,y_pred)
print(f'root mean sq error: {mse}')
```

```python
from sklearn.metrics import r2_score
# calculete r-squared
r2 = r2_score(Y_test,y_pred)
print(f'r-squaredis: {r2}')
```
